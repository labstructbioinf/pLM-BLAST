```python

import sys
sys.path.append('..')
import pandas as pd
import os

import torch
import matplotlib.pyplot as plt

from alntools.base import Extractor
import alntools as aln
```

```python
emb = torch.load('ctermialbug.pt')
with open('ctermialbug.fasta', 'rt') as fn:
    file = fn.readlines()
seq1 = file[1].replace('\n', '')
seq2 = file[3].replace('\n', '')
print(len(emb))
emb1 = emb[0].numpy()
emb2 = emb[1].numpy()
```

```python
file
```

```python
extr = Extractor()
extr.GAP_OPEN = 0.0
extr.WINDOW_SIZE = 10
```

```python
res12, density12, paths12, scorematrix12 = extr.embedding_to_span(emb1, emb2, mode='all')
paths12 = [p - 1 for p in paths12]
mask12 = aln.prepare.mask_like(densitymap=density12, paths=paths12)
coords = res12.indices.tolist()[1]
res12.sort_values(by=['len'], ascending=False).head(10)
```

```python
for idx, row in res12.iterrows():
    aln.draw_alignment(row.indices, seq2, seq1, output=None)
```
